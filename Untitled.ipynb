# 1. What is the distribution / values for owned/leased


```python
plt.bar(parking['Owned/Leased'], parking['Total Parking Spaces']) #plots owned/leased on x-axix and Total parking spaces on the Y-axis
```

```python
owned_leased= pd.pivot_table(parking, index = 'Owned/Leased', values = 'Total Parking Spaces', aggfunc = 'count') #creating a pivot table to see the values
owned_leased
```

```python
owned_leased.plot(kind= 'bar')
```

# 2 In which building state is parking situation in excess. Have it in order

```python
parking
```

```python
excess_parking = pd.pivot_table(parking, index = 'Bldg State', columns = 'Bldg Status', values = "Total Parking Spaces")
excess_parking
```

```python
excess_parking = excess_parking[excess_parking.EXCESS > 0]
excess_parking
```

```python
excess_parking.plot(kind= 'bar')
```

# 3. The bldg cities that have been decomissioned along with total parking

```python
decom_city = parking[parking['Bldg Status'] == 'DECOMMISSIONED']
decom_city
```

# In which property is the parking space in excess

```python
excess_parking_property = pd.pivot_table(parking[parking['Bldg Status']=='EXCESS'], index = 'Property Type', values = 'Total Parking Spaces')
excess_parking_property
```

# Where is more active parking- owned/leased 

```python
active_parking = pd.pivot_table(parking[parking['Bldg Status'] == 'ACTIVE'], index = 'Owned/Leased', values = 'Total Parking Spaces')
active_parking
```

# Which states have more leased parking

```python
state_parking = pd.pivot_table(parking[parking['Owned/Leased'] == 'LEASED'], index = 'Bldg State', values = 'Total Parking Spaces')
state_parking
```

```python
state_parking.sort_values('Total Parking Spaces', ascending = False)
```
